# 그래프의 다양한 그림 속성

```python
import matplotlib.pyplot as plt

plt.plot([10,20,30,40], 'r.', label ='circle')
plt.plot([40,30,20,10], 'g^' ,label ='triangle up')

plt.title('marker')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [10,20,30,40], '-.m*', [1,2,3,4], [50, 60, 70, 80], 'r. ', label='circle')
plt.plot([40, 30, 20, 10], 'g ^', label = 'triangle up')

plt.title('marker')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(5,4))

for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(1000), bins = 50, color='g', alpha=0.9)

print(plt.rcParams['figure.subplot.wspace'])
print(plt.rcParams['figure.subplot.hspace'])

plt.subplots_adjust(wspace=0.3, hspace=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot([10,20,30,40], [1, 4, 9, 16], c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("Style")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], "-bo", lw=3, ms=8, mec="skyblue", mew=2, mfc="y")
plt.title("Style")
plt.show()
```
